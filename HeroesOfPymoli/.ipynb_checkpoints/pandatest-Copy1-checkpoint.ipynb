## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#get unique item
num_uniqueitems = len(purchase_data["Item Name"].value_counts())
#num_uniqueitems
ave_price = round(purchase_data["Price"].mean(),2)
#ave_price  how to round this in 2 decimal
num_purchase = purchase_data["Purchase ID"].count()
#num_purchase
total_revenue = round((ave_price)*(num_purchase),2)
#total_revenue
```

```python
summary_df=pd.DataFrame([{"Number of Unique Items":num_uniqueitems,
                        "Average Price":ave_price,
                        "Number of Purchases":num_purchase,
                        "Total Revenue":total_revenue}])
summary_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
Count_Male_Players = purchase_data.loc[purchase_data["Gender"]=="Male",:]
#Count_Male_Players
Unique_Count_Male_Players= Count_Male_Players["SN"].nunique()
Unique_Count_Male_Players
percent_male_players =round(((Unique_Count_Male_Players) / (totalplayer))*100,2)
percent_male_players
```

```python
Count_Female_Players = purchase_data.loc[purchase_data["Gender"]=="Female",:]
#Count_Female_Players
Unique_Count_Female_Players= Count_Female_Players["SN"].nunique()
Unique_Count_Female_Players
percent_female_players = round((Unique_Count_Female_Players / totalplayer) * 100,2)
percent_female_players
```

```python
Count_Other_Players = purchase_data.loc[purchase_data["Gender"]=="Other / Non-Disclosed",:]
#Count_Other_Players
Unique_Count_Other_Players= Count_Other_Players["SN"].nunique()
Unique_Count_Other_Players
percent_other_players = (Unique_Count_Other_Players / totalplayer) * 100
percent_other_players
```

```python
Gender_df = pd.DataFrame({
    "Total Counts":[Unique_Count_Male_Players,Unique_Count_Female_Players,Unique_Count_Other_Players],
    "Percentage":[percent_male_players,percent_female_players,percent_other_players]
},index=['Male','Female','Other / Non-Disclosed'])
Gender_df
```

```python
#get average total purchage per person, use the unique count per gender
AveTotPurchPer=round((TPurchaseVal)/(Gender_df["Total Counts"]),2)
AveTotPurchPer
```

```python
#Creating a Summary DataFrame
gendersummary_df=pd.DataFrame({"Purchase Count":purchase_counts,
                    "Average Purchase Price":AveragePrice,
                     "Total Purchase Value":TPurchaseVal,
                     "Avg Total Purchase per Person":AveTotPurchPer})
gendersummary_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
print(purchase_data["Age"].max())
print(purchase_data["Age"].min())
```

```python
unique_df = purchase_data["SN"].unique()
#unique_df
```

```python
#create bins 
bins = [0,9.9,14.9,19.9,24.9,29.9,34.9,39.9,45]
#create labels
grp_labels=["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]
```

```python
#slice the data and place it into bins
pd.cut(purchase_data["Age"], bins, labels=grp_labels).head()
```

```python
#place the data series into new column inside the Dataframe
purchase_data["Age Group"]=pd.cut(purchase_data["Age"], bins, labels=grp_labels)
purchase_data.head()
```

```python
#create a grpby object
agegrp = purchase_data.groupby('Age Group')
print(agegrp['Age'].count())
```
