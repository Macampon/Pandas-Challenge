## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting

```python
#Get unique item
num_uniqueitems = len(purchase_data["Item Name"].value_counts())
#num_uniqueitems
ave_price = purchase_data["Price"].mean()
#ave_price  how to round this in 2 decimal
num_purchase = purchase_data["Purchase ID"].count()
#num_purchase
total_revenue = round((ave_price)*(num_purchase),2)
#total_revenue
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
print(purchase_data["Age"].max())
print(purchase_data["Age"].min())
```

```python
#create bins 
bins = [0,9.9,14.9,19.9,24.9,29.9,34.9,39.9,45]
#create labels
age_labels=["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]
```

```python
#slice the data and place it into bins
pd.cut(purchase_data["Age"], bins, labels=age_labels).head()
```

```python
#place the data series into new column inside the Dataframe
purchase_data["Age Group"]=pd.cut(purchase_data["Age"], bins, labels=age_labels,include_lowest=True)
purchase_data.head()
```

```python
#create a grpby object

agegrp= purchase_data.groupby('Age Group')['SN'].nunique()
agegrp
```

```python
#Compute for the Age Grps Percentage 
#totalplayer is the count of unique player

agegrp_percent = purchase_data.groupby('Age Group')['SN'].nunique()/totalplayer
agegrp_percent_df=pd.DataFrame(agegrp_percent)
agegrp_percent_df
```


Age Demographics OUTPUT

```python
# Merge Data and Create Summary 

merge_agegrp_df=pd.merge(agegrp,agegrp_percent_df,on="Age Group")
merge_agegrp_df.columns=["Total Count","Percentage of Players"]
merge_agegrp_df["Percentage of Players"]=merge_agegrp_df["Percentage of Players"].map("{:.2%}".format)
merge_agegrp_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Copy existing binning for purchasing analysis
purchase_df = purchase_data
purchase_df.head()
```

```python
#Create groupby df and display purchase count

grp_purchase = purchase_df.groupby("Age Group")
grp_purchase_df = pd.DataFrame(grp_purchase["Purchase ID"].count())
grp_purchase_df
```

```python
#Computes for Average Purchase Price

ave_purchase_price = grp_purchase["Price"].mean()
grp_ave_purprice_df =pd.DataFrame(ave_purchase_price)
grp_ave_purprice_df
```

```python
#Computes for Total Purchase Value

total_purchase_val = grp_purchase["Price"].sum()
grp_total_purval_df =pd.DataFrame(total_purchase_val)
grp_total_purval_df
```

```python
#Computes for Average Total Purchase per Person

avg_tot_purchperson = total_purchase_val/purchase_data.groupby('Age Group')['SN'].nunique()
grpavg_tot_purchperson_df = pd.DataFrame(avg_tot_purchperson)
grpavg_tot_purchperson_df
```


MERGING OF DF's AND SUMMARY OUTPUT

```python
#Merge Purchage Count and Purchase Price 

merge_grp_avepurchase_df=pd.merge(grp_purchase_df,grp_ave_purprice_df,on="Age Group")
merge_grp_avepurchase_df.columns=["Purchase Count","Average Purchase Price"]
merge_grp_avepurchase_df["Average Purchase Price"]=merge_grp_avepurchase_df["Average Purchase Price"].astype("float").round(2)

merge_grp_avepurchase_df
```

```python
#Merge Total Purchase Value to the previous DF

merge_total_purval_df=pd.merge(merge_grp_avepurchase_df,grp_total_purval_df,on="Age Group")
merge_total_purval_df.columns=["Purchase Count","Average Purchase Price","Total Purchase Value"]
merge_total_purval_df
```

```python
#Merge the Avg Total Purcahse per Person and Create a Summary

summary_purchasing_df = pd.merge(merge_total_purval_df,grpavg_tot_purchperson_df,on="Age Group")
summary_purchasing_df.columns=["Purchase Count","Average Purchase Price","Total Purchase Value","Avg Total Purchase per Person"]
summary_purchasing_df["Avg Total Purchase per Person"] = summary_purchasing_df["Avg Total Purchase per Person"].astype("float").round(2)
summary_purchasing_df
summary_purchasing_df.style.format({"Average Purchase Price":"${:,.2f}", "Total Purchase Value":"${:,.2f}","Avg Total Purchase per Person":"${:,.2f}"})
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Retrieve columns from purchase_df
columns_popular = purchase_df[["Item ID","Item Name","Price"]]
#columns_popular
```

```python
#Create Grp
grp_poplar_itemid = columns_popular.groupby(["Item ID", "Item Name"])
#grp_poplar_itemid.head()
```

```python
purchase_count = grp_poplar_itemid["Price"].count()
tot_purval = grp_poplar_itemid["Price"].sum()
item_price = tot_purval / purchase_count
grp_popular_items = pd.DataFrame({"Purchase Count":purchase_count, 
                                  "Item Price" : item_price,
                                  "Total Purchase Value" : tot_purval
                                 })
grp_popular_items.head()
```


MOST POPULAR ITEMS OUTPUT

```python
#SORT Per Purchase Count 
grp_popular_items_sort = grp_popular_items.sort_values("Purchase Count", ascending=False).head()
#grp_popular_items_sort.head()
grp_popular_items_sort.style.format({"Item Price":"${:,.2f}", "Total Purchase Value":"${:,.2f}"})
```
